```python
import os
```

**Import pyspark related libraries**

```python
import pyspark
from pyspark.sql.session import SparkSession
```

**Set pyspark Java 8 JDK location**

```python
java8_location = '/usr/lib/jvm/java-8-openjdk-amd64'  # Set your own
os.environ['JAVA_HOME'] = java8_location
```

**Initialize pyspark**

```python
app = SparkSession.builder.appName("A1").getOrCreate()
sc = app.sparkContext
spark = SparkSession(sc)
```

```python
csv_folder = "../data/"
csv = "datadump.csv"
df = spark.read.csv(csv_folder + csv, header=True)
df.show()
```

```python
df.count()
```
